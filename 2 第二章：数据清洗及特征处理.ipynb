### 2.3.1 任务一：对年龄进行分箱（离散化）处理

```python
df.head()
```

```python
ages_cut = pd.cut(df['Age'],bins = 5)
ages_cut   # 默认 right = True
```

```python
ages_cut_1 = pd.cut(df['Age'],bins = 5,labels = [1,2,3,4,5])
ages_cut_1
```

```python
ages_cut_2 = pd.cut(df['Age'],bins = [0,5,15,30,50,80])
ages_cut_2
```

```python
ages_cut_3 = pd.cut(df['Age'],bins = [0,5,15,30,50,80],labels = [1,2,3,4,5])
ages_cut_3
```

```python
ages_cut_4 = pd.cut(df['Age'],bins = [0,5,15,30,50,80],include_lowest=True,right=False)
ages_cut_4 # 若设定 include_lowest = True ,则必须right = False
```

```python
ages_cut_5 = pd.qcut(df['Age'],5)
ages_cut_5
```

```python
df['ages_band'] = ages_cut_5
df.head()
```
